 #  create function to generate Data set like 

```python
#set seed
np.seed(111)
# Function to generate test data

def CreateDataSet(num_range = 1):
    output = []
    for i in range(num_range):
        
        #create a weekly(mondays) data range
        date_rng = panda.date_range(start='1/1/2009', end='12/31/2012', freq='W-MON')
        
        # create random data which is equal in size with date_range
        data = np.randint(low=25,high =1000,size=len(date_rng))
        
        ## status pool
        status = [1,2,3]
        
        # make random list of status of 1 or 2 or 3 
        # total random status = lenght of date_rng
        random_status = [status[np.randint(low=0,high=len(status))] for i in range(len(date_rng))]
        
        # State pool
        states = ['BKT','KTM','LALIT','bkt','BRJ','LUM']
        
        # Make a random list of states
        
        random_states = [states[np.randint(low=0,high=len(states))] for i in range(len(date_rng))]
        
        # now zip all random_states , random_Status, data , date_rng  and append to the extend
        output.extend(zip(random_states,random_status,data,date_rng))
    
    return output
        
    
```

 We have funtion called CreateDataSet to generate test data and work it out

```python
dataset = CreateDataSet(4)#pass the number of range
df = panda.DataFrame(data=dataset, columns=['State','Status','CustomerCount','StatusDate'])
df.info()
```

Our task is now to create a new dataframe that compresses the data so we have daily customer counts per State and StatusDate. We can ignore the Status column since all the values in this column are of value *1*. To accomplish this we will use the dataframe's functions ***groupby*** and ***sum()***.  

Note that we had to use **reset_index** . If we did not, we would not have been able to group by both the State and the StatusDate since the groupby function expects only columns as inputs. The **reset_index** function will bring the index ***StatusDate*** back to a column in the dataframe. 

```python
# Group by State and StatusDate
df.reset_index().head()
```

```python
df.reset_index().groupby
```

```python
a= df.sum()# add all the status and customer count and state
b = df.reset_index().sum()
c = df.reset_index().groupby(['CustomerCount','State']).sum()
d = df.reset_index().groupby(['State','StatusDate']).sum()
```
